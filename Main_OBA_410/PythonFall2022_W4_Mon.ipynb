# Querying DataFrames

```python
# boolean mask to query DataFrames
# get patients who have been pregnant more than 2 times
df2['Pregnancies']>2
```

```python
# we pass boolean masks to our data using []
df2[df2['Pregnancies']>2]
```

```python
# get the patiesnts that at most 32 years -- age<=32
# can store the reslts in new variable if we want to
atmost32=df2['Age']<=32
atmost32
```

```python
df2[atmost32]
```

```python
# pass the condition directly
# same as what de just did; passes it directly into boolean mask
df2[df2['Age']<=32]
```

```python
# get patients who are younger than 32-- age<32
younger32=df2['Age']<32
df2[younger32]
```

```python
# patients with pregnancies not equal to 1 (either 0 or more than 1)
# to do "not equal to", use the != operator
pregNot1=(df2['Pregnancies']!=1)
df2[pregNot1]
```

```python
# patients with BP at least 64 AND being pregnant more than once
# can query more than one variable at a time
BPatleast64=df2['BloodPressure']>=64
Pregmore1=df2['Pregnancies']>1
df2[BPatleast64 & Pregmore1]
```

```python
df2[(df2['BloodPressure']>=64)&(df2['Pregnancies']>1)]
```

```python
## in the original df
# the number of patiets patients with BP at least 64 AND being pregnant more than once
# must use .shape to see how many there are actually in the data set
BPatleast64=df['BloodPressure']>=64
Pregmore1=df['Pregnancies']>1
df[BPatleast64 & Pregmore1].shape
```

```python
# can also use the len function instead of .shape to find number of rows
len(df[BPatleast64 & Pregmore1])
```

```python
# back to df2
# patients with BMI more than 30 OR being pregnant exactly once
# OR sing (|), shift+baclslash (right above Enter key) (MUST DO OR THIS WAY)
BMImore30=df2['BMI']>30
preg1=(df2['Pregnancies']==1)
df2[BMImore30 | preg1]
```

```python
# paranthsese matter!
# patients (younger than 32 AND BP above 65) OR Glucose is 148
# HOW TO USE AN "AND" AND AN "OR" IN THE SAME STATEMENT
con1=df2['Age']<32
con2=df2['BloodPressure']>65
con3=df2['Glucose']==148
df2[(con1 & con2) | con3]
```

```python
# patients younger than 32 AND (BP above 65 OR Glucose is 148)
df2[con1 & (con2 | con3)]
```

# Handling missing values

```python
# read diabetes_w_missing data
df_miss=pd.read_csv('diabetes_w_missing-1.csv')
df_miss
```

We can handle missing values in two ways:
    
    1-- drop them (either rows or columns)
    
    2-- fill them (impute them)

## dropping missing values

```python
# getting info on dataset
# look at non-null count
df_miss.info()
```

```python
# describing dataset
## use count row to tell if there are any missing values
df_miss.describe()
```

```python
# drop rows with missing (the default)
df_miss.dropna()
```

```python
# dropping columns with missing (not the default, must change the axis option)
# if we want make the drops permanent, we set inplace to True
df_miss.dropna(axis=1)
```

```python
# because we did not make it permanent, the DataFrame goes back to normal
df_miss
```

## filling or imputing missing values

```python
# mean values of columns
df_miss.mean()
```

```python
# filling missing values with mean
df_miss.fillna(value=df_miss.mean(),inplace=True)
df_miss
```

```python
# mode to fill categorical variables
df_miss['Gender'].mode()
```

```python
df_miss['Gender'].mode()[0]
```

```python
# dictionary to specify the column lables and values to use for filling
df_miss.fillna(value={'Gender':'Female'},inplace=True)
df_miss
```
