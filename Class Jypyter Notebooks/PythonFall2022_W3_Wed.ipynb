# Renaming

```python
# create a new DataFrame as the first 5 rows of df
# make sure to add copy() at the end
df1 = df.iloc[:5].copy()
df1
```

```python
# renaming columns and indexes
df1.rename(columns={"Pregnancies":"preg_num", "BloodPressure":"BP"},
          index={0:"p1", 1:"p2", 2:"p3", 3:"p4", 4:"p5"},inplace=True)
```

# Dropping Rows and Cols

```python
# drop p1,p3
df1.drop(["p1","p3"])
```

```python
# drop columns -- drop BMI
df1.drop(columns=["BMI"])
```

```python
# drop 01, p2, and BP and Glucose
df1.drop(index=["p1","p2"], columns = ["BP", "Glucose"])
```

# Adding new Rows and Cols

## New Col

```python
# add a new col and assign value 5 for all rows for that col
df1["new_col1"]=5
df1
```

```python
# add a new col with different values for each row
df1["new_col2"]=[11,4,6,0,-1]
df1
```

```python
# new col, assigning values using the values of existing cols
# newcol3, values are summation of new_col1 and newcol2
df1["new_col3"]= df1["new_col1"]+df1["new_col2"]
df1
```

## Adding New Rows

```python
# we have to use loc
df1.loc["p6"]=None
df1
```

```python
import numpy as np
```

```python
# means the same as above
# we have to use loc
df1.loc["p6"]=np.nan
df1
```

```python
# new row using an existing row
# new row p7, use the value from p2
df1.loc["p7"]=df1.loc["p2"]
df1
```

```python
# Deleting Duplicates
# to make it permanet, set inplace to True
df1.drop_duplicates()
```
