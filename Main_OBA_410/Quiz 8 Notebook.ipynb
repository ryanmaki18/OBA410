```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# reading the data
cancer=pd.read_csv('breast_cancer_data.csv',index_col=0)
cancer.head(2)
```

```python
# creating features and target
X_cancer, y_cancer=cancer.iloc[:,1:], cancer['diagnosis']
X_cancer.head(2)
```

```python
# split to train and test
X_train, X_test, y_train, y_test=train_test_split(X_cancer, y_cancer, random_state=0)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# define and fit the model
rf1=RandomForestClassifier(n_estimators=200, random_state=3)
rf1.fit(X_train, y_train)
```

```python
# evaluate the performance
print('rf1 on train: {:.2%}'.format(rf1.score(X_train, y_train)))
print('rf1 on test: {:.2%}'.format(rf1.score(X_test, y_test)))
```

```python
rf1.feature_importances_
```

```python
rf_feat_imp=pd.DataFrame(data=rf1.feature_importances_, index=X_cancer.columns, columns=['importance'])
rf_feat_imp.sort_values(by=['importance'], ascending=False) 
```
