```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import cross_val_score, train_test_split
```

# Cross Validation

## Breast Cancer Data

```python
# reading the data
cancer = pd.read_csv("breast_cancer_data-1.csv", index_col=0)
cancer.head(5)
```

```python
# creating features and target sets
# diagnosis is target, everything else is a feature
x_cancer, y_cancer = cancer.iloc[:,1:], cancer['diagnosis']    # for target set, we want it 1-dim
# checking our work
display(x_cancer.head(3))
display(y_cancer.head(3))
```

```python
# define the model
knn_cls_cancer = KNeighborsClassifier(n_neighbors = 5)
```

```python
# cross-validation
scores = cross_val_score(knn_cls_cancer, x_cancer, y_cancer, cv=6)
scores
```

```python
# the average test score on 6-folds
scores.mean()
```

```python
print('Average test accuracy in a 6-fold cross val is: {:.2%}'.format(scores.mean()))
```

## Boston Housing Data

```python
# reading the data
boston = pd.read_csv("boston_housing_data-1.csv", index_col=0)
boston.head(5)
```

```python
# For Regression Problems, before doing cross validation, we always shuffle our data
boston = boston.sample(frac=1,random_state=0)
```

```python
# creating features and target sets
# target_medv is target, everything else is a feature
x_boston, y_boston = boston.iloc[:,:-1], boston['target_medv']
display(x_boston.head(3))
display(y_boston.head(3))
```

```python
# define the model with 5 neighbors
knn_reg_boston = KNeighborsRegressor(n_neighbors = 5)
```

```python
# cross-validation with 5 folds
scores1 = cross_val_score(knn_reg_boston, x_boston, boston, cv=5)
scores1
```

```python
print('Average test accuracy in a 5-fold cross val is: {:.2%}'.format(scores1.mean()))
```
